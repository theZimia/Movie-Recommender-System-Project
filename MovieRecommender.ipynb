```python
import ast

def convert(obj):
    L = []

    # Case 1: already a list → just return it
    if isinstance(obj, list):
        return obj

    # Case 2: string → convert
    try:
        for i in ast.literal_eval(obj):
            L.append(i['name'])
    except:
        return []

    return L

movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    try:
        # If string → convert
        if isinstance(obj, str):
            obj = ast.literal_eval(obj)

        # If already list of names → just slice
        if isinstance(obj, list):
            # Case: list of dicts
            if len(obj) > 0 and isinstance(obj[0], dict):
                return [i['name'] for i in obj[:3]]
            # Case: list of strings (already processed)
            else:
                return obj[:3]

        return []

    except (ValueError, SyntaxError, TypeError, KeyError):
        return []
```

```python
movies['cast']= movies['cast'].apply(convert3) 
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':      
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
vectors
```

```python
stem('In the 22nd century, a paraplegic Marine is ')
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
